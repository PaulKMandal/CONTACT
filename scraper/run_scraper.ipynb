```python
from scrape_utils import wayback_scraper
import datetime
#test = wayback_scraper("https://www.foxnews.com")
test = wayback_scraper("https://www.aljazeera.com")
words= ["ISIL","ISIS","war"]
```

```python
date = datetime.datetime(2013,1,1)
days_to_parse = 100
for i in range(days_to_parse):
    test.set_date(y=date.year,m=date.month,d=date.day)
    query = test.search_by_title(words)

    for article in query:
        print("Current Date:",date)
        print(article.publish_date)
        print(article.title)
        print(article.url)
    
    date+=datetime.timedelta(days=1)
    
```
